## Installation
* We need to install the **cfclon** package
  and harmonic matrices database needed for this demo.

```python
%pip install -q https://samyzaf.com/cfclon-33.zip
from cfclon import *
plt.rcParams['figure.dpi'] = 90
reg.log.state = "disabled"
dbname = "https://samyzaf.com/cfproj/aleph2a.db"
load_db(dbname)
```

```python
url = "https://samyzaf.com/cfproj/aleph2.svg"
aleph = read_polygon(url, tag="aleph")
cage = read_polygon(url, tag="cage")
draw_cage(cage)
draw_curve(aleph, facecolor="aliceblue", fill=True)
show_plot()
```

* In the next code we clone the harmonic coordinates of
  the aleph figure into a new cage, which you can
  download from the following url.

```python
url = "https://samyzaf.com/cfproj/aleph2b.svg"
aleph = read_polygon(url, tag="aleph")
cage2 = read_polygon(url, tag="cage2")
draw_cage(cage2)
coords_list = harmonic_coordinates(aleph)
cloned_points = clone(coords_list, cage2)
draw_curve(cloned_points, fill=True, linewidth=1, facecolor="cyan")
show_plot()
```

* You can experiment with different cages.
* We defined a new cage (`cage3`), draw it, and then
  clone our Aleph figure in it.

```python
cage3 = [(167.9, 739.0), (728.0, 409.0), (352.0, 59.0), (67.9, 23.0)]
draw_cage(cage3)
cloned_points = clone(coords_list, cage3)
draw_curve(cloned_points, fill=True, linewidth=1, facecolor="papayawhip")
show_plot()
```

Lets see how it looks using Mean Value coordinates:

```python
mean_coords = get_mean_value_coords_map(cage)
coords_list = mean_coords(aleph)
cloned_points = clone(coords_list, cage3)
draw_cage(cage3)
draw_curve(cloned_points, fill=True, linewidth=1, facecolor="cornsilk")
show_plot()
```

* This is almost identical to what we got using Harmonic Coordinates!
* It shows that in many cases **Mean Value Coordinates** are OK to use.
* They are much cheaper computationaly and 100%
  precise (we have closed form formula for the in contrast to an approximate numeric computation for Harmonic Coordinates).
