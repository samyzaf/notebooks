## The **man** character
* The animation character that we use for this project is
  named **man**, as it is a simple 2D man figure.
* This is a large polygon with 178 vertices!
* So the only way to define and hold this data is by using
  a graphics file.
* We use the [**SVG**](https://en.wikipedia.org/wiki/SVG)
  graphics file format.
* To manipulate this figure we use a 21 vertices cage which
  we call **cage1**.
* This file also contains the definition of a second cage
  which we call **cage2**.
* In the following code we load the coordinates of our **man**,
  **cage1**, and **cage2** figures from a **url** address, and assign
  them to our **Python** variables.
* There is no need to know anything about **SVG** for this project
  except this.

```python
svg = "https://samyzaf.com/cfproj/man.svg"
man = read_polygon(svg, tag="man")
cage1 = read_polygon(svg, tag="cage1")
cage2 = read_polygon(svg, tag="cage2")
draw_curve(man, fill=True, facecolor="aliceblue", linewidth=0.5)
draw_cage(cage1, linewidth=0.5)

show_plot()
```

## Compute the harmonic coordinates of the **man** polygon
* As seen in many examples to compute the
  harmonic coordinates of all polygon points,
  we use the command  
  **harmonic_coordinates(polygon)**

* It returns a list of harmonic coordinates of each
  polygon point.

```python
man_coords_list = harmonic_coordinates(man)
```

* This is the list of the 178 vectors, where each vector has 21 entries!
* So the **man_coords_list** consist of $178\times 21 = 3738$
  real numbers!
* To view it as table, you can run the following code.
* (This is a very large table, so it's not recommended..)

```python
import pandas
data = pandas.DataFrame(man_coords_list)
data
```
