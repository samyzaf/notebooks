# **Solving a first order linear equation by FDM**

<small>version: 08.10.2025e</small>

* We show how to find a numerical solution to
  the ordinary differential equation  
  $\qquad
  \left\{
  \begin{array}{l}
  y'  = 4y + 3x, \quad 0<x<2
  \\[1ex]
  y(0) = -2
  \end{array}
  \right.
  $  

* Our interval $[0,2]$ is divided to $n=100$ equal subintervals
  with nodes $x_i = i\Delta x$, where $\Delta x=2/100$.

* We use the forward difference derivative formula
  $y'=\frac{y_{i+1} - y_{i}}{\Delta x}$.

* Hence  
  $\qquad\displaystyle
  \frac{y_{i+1} - y_i}{\Delta x} = 4y_i + 3x_i, \quad i=0,1,2,\ldots,n-1
  $

* From which we immediately get a recursive solution  
  $\qquad\displaystyle
  \left\{
  \begin{array}{l}
  y_{i+1} = y_i + \left(4y_i + 3x_i\right) \Delta x, \quad i=0,1,2,\ldots,n-1
  \\
  y(0) = -2
  \end{array}
  \right.
  $

* This is easy to code:

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.dpi'] = 100
plt.figure(figsize=(8, 4))

N = 100
dx = 2/N
x = np.array([i*dx for i in range(N+1)])
y = np.zeros(101)
y[0] = -2
for i in range(0,N):
    y[i+1] = y[i] + dx*(4*y[i] + 3*x[i])

plt.plot(x,y)
plt.show()
```

  our differential equation!
