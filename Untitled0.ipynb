```python
import numpy as np
from numpy import meshgrid, linalg, array, linspace, arange, vectorize
import matplotlib.pyplot as plt
from matplotlib import ticker, colors
from matplotlib.ticker import LinearLocator, MultipleLocator, FormatStrFormatter, AutoMinorLocator
from matplotlib import cm
```

```python
def chi(x,y):
    if x>=y:
        return 1
    return 0

def u(x,y):
    z = 5 * (y-x)**2 * chi(x,y) - 5*y**2
    return z

```

```python
fig = plt.figure(figsize=(5,5), dpi=120)
ax = fig.add_subplot(111, projection='3d')
a = 8
b = 8
dx = dy = 0.05
X = arange(0, a, dx)
Y = arange(0, b, dy)
X,Y = meshgrid(X, Y)
vec_u = vectorize(u)
Z = vec_u(X,Y)
cmap = plt.cm.jet
#surf = ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10, cmap=cmap, linewidth=1.0, antialiased=True)
surf = ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap=cmap, linewidth=0.4, edgecolors='0.3', alpha=0.8, antialiased=True)
ax.view_init(elev=40, azim=25, roll=0)
plt.tight_layout()
#plt.savefig('surf1.jpg', bbox_inches='tight')
plt.show()
```
